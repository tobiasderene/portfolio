# Introduccion

## ¿Que son los datos abiertos?

Los datos abiertos son información y conjuntos de datos que están disponibles para cualquier persona, sin restricciones de acceso, uso o redistribución. Estos datos son generalmente proporcionados por entidades gubernamentales, organizaciones públicas y privadas, y pueden abarcar una amplia variedad de temas, como estadísticas demográficas, información geográfica, datos climáticos, resultados de investigaciones, presupuestos gubernamentales, entre otros.

## ¿Que tiene que ver con el Paraguay?

En el contexto de Paraguay, los datos abiertos son importantes por varias razones. En primer lugar, promueven la transparencia y la rendición de cuentas en el gobierno, permitiendo a los ciudadanos acceder a información clave sobre cómo se toman decisiones y se utilizan los recursos públicos. Esto ayuda a fortalecer la democracia y fomenta la participación ciudadana en los asuntos públicos.

## ¿Que aporte tienen los datos abiertos  la sociedad?

Los aportes que ofrecen son varios, empezando con la transparencia ya que los datos abiertos permiten a los ciudadanos tener acceso a información confiable y actualizada sobre las políticas públicas, los servicios gubernamentales y otros aspectos relevantes.

Por otro lado es una potente herramienta para fomentar la innovación y el desarrollo de aplicaciones, Los emprendedores y desarrolladores pueden utilizar estos datos para crear soluciones que aborden desafíos sociales, mejoren la eficiencia de los procesos y brinden beneficios tangibles a la sociedad.

Asi mismo se puede mejorar el servicio publico, utilizando los datos podemos ver en que sectores de los servicios publicos hay deficiencias y buscar la forma de solucionarlo.

### Limpiar Datos en serio 

Una vez listo todos los pasos anteriores, es en este momento en el cual busco entre los datos (especialmente Strings) que son cosas las cuales pueden molestarme al analizar.

Un problema que tiene esta base son los codigos dentro de las descripciones, ya no son un elemento necesario para mi, por lo que podemos utilizar la funcion .split() para separar los datos por un elemento en esto y convertirlo en un array con la cantidad de reglones en nuestro texto +1, como los datos relevantes se encuentran en el segundo elemento de la lista (ejemplo: [1 , PROGRAMA CENTRAL]) se pone el valor [1] al final de esta linea para que se eliga solamente la descripcion

```python
df_pres['funcion'] = df_pres['funcion'].str.split(' - ').str[1]
df_pres['descripcionNivel'] = df_pres['descripcionNivel'].str.split('-').str[1]
df_pres['descripcionEntidad'] = df_pres['descripcionEntidad'].str.split('-').str[1]
df_pres['descripcionUnidadResponsable'] = df_pres['descripcionUnidadResponsable'].str.split(' - ').str[1]
df_pres['descripcionTipoPresupuesto_ClasePrograma'] = df_pres['descripcionTipoPresupuesto_ClasePrograma'].str.split('-').str[1]
df_pres['descripcionPrograma'] = df_pres['descripcionPrograma'].str.split(' - ').str[1]
df_pres['descripcionSubprograma'] = df_pres['descripcionSubprograma'].str.split(' - ').str[1]
df_pres['descripcionProyecto_Actividad'] = df_pres['descripcionProyecto_Actividad'].str.split('-').str[1]
df_pres['finalidad'] = df_pres['finalidad'].str.split(' - ').str[1]
df_pres['subfuncion'] = df_pres['subfuncion'].str.split('-').str[1]
df_pres.head()
```

### ¿Que tan representante es el Programa Central en ambos set de datos?

Para saber esto primero realizaremos los calculos correspondientes a cada DF y luego lo graficaremos para tener una mejor nocion de cuan grande es

```python
pres_sum = df_pres.groupby('descripcionPrograma')['montoVigente'].sum().sort_values(ascending=False)
dic_pres_sum = pres_sum.to_dict() 
pc = dic_pres_sum.get('PROGRAMA CENTRAL')
tot = df_pres['montoVigente'].sum()
print(f'El programa central representa un {(pc/tot)*100}%  de la suma de los datos del PGN')
```

```python
nom_sum = df_nom.groupby('descripcionPrograma')['montoPresupuestado'].sum().sort_values(ascending=False)
dic_nom_sum = nom_sum.to_dict() 
pc = dic_nom_sum.get('PROGRAMA CENTRAL')
tot = df_nom['montoPresupuestado'].sum()
print(f'El programa central representa un {(pc/tot)*100}%  de la suma de los datos del PGN')
```

### ¿Cuanto representa el salario de las nominas en el PGN filtrado por el Programa Central en ambos Sets?

Despues de un largo de proceso llegamos a la fase final, donde solamente necesitamos filtrar, conseguir el porcentaje que representa los salarios de la nomina y compararlos con los mayores gastos del Programa Central 

Empezaremos calculando la representacion del salario

```python
df_presProgramaCentral = df_pres.loc[df_pres['descripcionPrograma'].isin(['PROGRAMA CENTRAL'])]
df_nomProgramaCentral = df_nom.loc[df_nom['descripcionPrograma'].isin(['PROGRAMA CENTRAL'])]
```

```python
pd.set_option('display.float_format', lambda x: '%.3f' % x)
totalPresPC = df_presProgramaCentral['montoVigente'].sum()
totalNomPC = df_nomProgramaCentral['montoPresupuestado'].sum()
porcentaje = (totalNomPC/totalPresPC)*100
print(f'El monto vigente de salarios en la Nomina de empleados representa un {porcentaje}% del total de gastos en el Programa Central del PGN\nLa cifra exacta es de {totalNomPC}')
```

Ahora buscaremos el top de los finalidades con mas gastos en el PGN y lo seteamos en un dictionary

```python
dic_top = df_presProgramaCentral.groupby('funcion').sum()['montoVigente'].sort_values(ascending=False).nlargest(n=5).to_dict()
dic_top.update({'SUELDOS':totalNomPC})
```

Por ultimo graficamos todos estos para ver la comparacion

```python
x = list(dic_top.keys())
y = list(dic_top.values())

plt.figure(figsize=(12, 10))
plt.barh(x, y)
plt.title('Gráfico de Barras')
plt.xlabel('Finalidades')
plt.ylabel('Datos')
plt.show()
```
